### 다층 퍼셉트론 신경망

걸음 수, 이동 거리, 심박 수를 입력하면 칼로리 소모량을 예측하는 모델을 개발하며 다층 퍼셉트론을 이해해보자.

다음과 같이 입출력을 구성한다.
- X
    - 걸음 수
    - 이동 거리
    - 정적인 시간(분)
    - 약간 활동적인 시간(분)
    - 상당히 활동적인 시간(분)
    - 매우 활동적인 시간(분)
- Y
    - 칼로리 소모량
    
다음과 같이 데이터셋을 구성한다.
- training set: df[:100]
- test set: df[100:]

```python
import pandas as pd

df = pd.read_csv('fitbit_export_2021.csv', thousands = ',')
print(df.columns)
```

```python
x = df[['걸음 수', '이동 거리', '정적인 시간(분)', '약간 활동적인 시간(분)',
       '상당히 활동적인 시간(분)', '매우 활동적인 시간(분)']]
y = df['칼로리 소모량']

x_train = x[:100]
y_train = y[:100]
x_test = x[100:]
y_test = y[100:]
```

모델 구성에 앞서 딥러닝 모델이 잘 학습할 수 있도록 입력 데이터의 전처리를 수행한다. 
속성별로 평균을 구한 뒤 각 항목에 평균을 빼고 표준편차로 나눠서 데이터 스케일링해야 단위나 범위에 따라 값이 다르게 나오는 경우를 방지할 수 있다.

train set, test set 둘 다 적용해야 한다.

```python
mean = x_train.mean(axis=0)
x_train -= mean

std = x_train.std(axis=0)
x_train /= std

print(mean, std)
```

input feature가 6개이고 output이 1개인 네트워크 아키텍처를 설계한다. 
비선형 문제를 풀 수 있도록 'relu' 활성화 함수를 가진 여러 층으로 구성했다.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(16, input_dim=6, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='linear'),
    
])

model.summary()
```

```python
'''
1. 모델 컴파일
- 손실 함수: mse
- optimizer: rmsprop
- 평가 matric: mae
2. 모델 학습
'''
model.compile(loss='mse', optimizer='rmsprop', metrics=['mae'])
model.fit(x_train, y_train, epochs=1000, batch_size=4)
```

```python
'''
3. 모델 예측
4. 모델 평가(정확도)
'''
x_test = (x_test - mean) / std
y_pred = model.predict(x_test)
print(y_pred)

loss, mae = model.evaluate(x_test, y_test)
print(mae)
```

```python
'''
제출
'''
import numpy as np

np.savetxt('y_pred.csv', y_pred, fmt='%f')
```
