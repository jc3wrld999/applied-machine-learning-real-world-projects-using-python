### 단층 퍼셉트론 신경망
먼저 걸음 수를 입력하면 이동 거리를 예측하는 간단한 모델을 만들면서 단층 퍼셉트론을 이해해보자.

다음과 같이 모델 입출력을 구성한다.
- X: 걸음 수
- Y: 이동 거리

다음과 같이 데이터셋을 구성한다.
- training set: df[:100]
- test set: df[100:]


```python
x = df['걸음 수']
y = df['이동 거리']

x_train = x[:100]
y_train = y[:100]
x_test = x[100:]
y_test = y[100:]
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_train, y_train)
plt.xlabel('steps')
plt.ylabel('distance')
plt.show()
```

걸음수와 이동거리는 선형적인 관계를 가진다. 이런 패턴은 선형회귀로도 풀 수 있고 여러가지 모델로 풀 수 있지만 가장 간단한 단층 퍼셉트론 신경망으로 풀어보자.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

Y = w * X + b 에서 주어진 X, Y에 대해 학습에 의해 갱신될 가중치(w) 하나와 바이어스(b)를 하나를 가진 네트워크 아키텍처를 정의한다.

```python
model = Sequential([
    Dense(1, input_dim=1)
])
model.summary()
```

손실 함수는 기계 학습 모델에서 모델의 예측값과 실제 값 사이의 차이를 측정하는 함수다. 

모델의 학습 과정에서 손실 함수를 최소화하는 방향으로 모델의 파라미터를 조정하게 된다.
- 평균 제곱 오차(Mean Squared Error, MSE): 회귀(Regression) 문제에서 널리 사용되는 손실 함수로 예측값과 실제 값의 차이의 제곱의 평균을 계산한다.

손실 함수는 회귀문제에 적합한 mse로 설정하고 optimizer는 rmsprop으로 설정하여 모델을 컴파일한다. 평가 매트릭은 수치 차이를 직관적으로 알 수 있도록
