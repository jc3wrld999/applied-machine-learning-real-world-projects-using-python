문장을 입력해서 이진 분류하는 모델에 대해서 알아보자.

언어가 시계열적인 의미가 있으므로 이 언어를 문자로 표현한 문장도 시계열적인 의미가 있다.

모델에 입력하기 위해 문장을 시계열 수치로 인코딩하는 방법과 이진 분류 모델을 구성해보고 학습 결과를 살펴보자.

이 모델들은 문장 혹은 시계열 수치로 양성/음성을 분류하거나 이벤트 발생 유무를 감지하는 문제를 풀 수 있다.

````python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Embedding, LSTM, Flatten
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.utils import plot_model
from tensorflow.keras.datasets import imdb
````

````python
'''
데이터셋 준비(IMDB)
'''
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=20000) # 20000번째로 많이 사용하는 단어까지만 데이터셋 만들기
print(x_test)
````

````python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
````

리뷰의 길이가 다르면 샘플의 길이가 달라진다. 모델의 입력으로 사용하기 위해선 고정된 길이로 변환해야 하기 때문에 keras에서 제공하는 전처리 함수인 sequence의 pad_sequences() 함수를 사용한다. 이 함수는 두가지 역할을 수행한다.
- 문장의 maxlen 인자로 지정된 길이로 맞춰준다. 예를 들어 200으로 지정했다면 200보다 짧은 문장은 0으로 채우고 200보다 긴 문장은 200단어까지 자른다.
- (num_samples, num_timesteps)으로 2차원 numpy 배열로 만들어준다. maxlen을 200으로 지정했다면 num_timesteps도 200이 된다.

````python
# from keras.preprocessing import sequence
from keras.utils.data_utils import pad_sequences

x_train = pad_sequences(x_train, maxlen=200)
x_test = pad_sequences(x_test, maxlen=200)
````

````python
x_train.shape
````

### 다층 퍼셉트론 신경망 모델

임베딩(Embedding) 레이어로 단어들을 인코딩한 후 Dense 레이어를 통해 분류하는 다층 퍼셉트론 신경망 모델이다. 

```python
model = Sequential()
model.add(Embedding(20000, 128, input_length=200)
```

임베딩 레이어의 매개변수
- input_dim: 단어 사전의 크기. 여기선 20000개의 단어 종류가 있다는 것을 의미. 이 값은 앞서 imdb.load_data() 함수의 num_words 인자값과 동일해야한다.
- output_dim: 단어를 인코딩한 후 나오는 벡터의 크기이다. 
    - 이 값이 128이라면 단어를 128차원의 의미론적 기하공간에 나타낸다는 의미이다. 
    - 단순하게 빈도수만으로 단어를 표시한다면 10과 11은 빈도수는 비슷하지만 단어로 볼 때는 전혀 다른 의미를 가지고 있다. 
    - 하지만 의미론적 기하공간에서는 거리가 가까운 두 단어는 의미도 유사하다.
    - 즉 임베딩 레이어는 입력되는 단어를 의미론적으로 잘 설계된 공간에 위치시켜 벡터로 수치화 시킨다고 볼 수 있다.
- input_length: 단어의 수, 즉 문장의 길이를 나타낸다. 
    - 임베딩 레이어의 출력 크기는 샘플 수 * output_dim * input_length가 된다.
    - 임베딩 레이어 다음에 Flatten 레이어가 온다면 반드시 input_length를 지정해야 한다.
    - Flatten 레이어인 경우 입력의 크기를 알아야 이를 1차원으로 만들어서 Dense 레이어에 전달할 수 있기 때문이다.

````python
'''
1. 네트워크 아키텍쳐 정의
'''
model = Sequential()
model.add(Embedding(20000, 128, input_length=200))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
````

````python
'''
2. 컴파일
3. 학습
'''
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

checkpoint_callback = ModelCheckpoint('best_model.h5', save_best_only=True, monitor='val_loss')

hist = model.fit(x_train, y_train, epochs=5, batch_size=64, validation_split=0.2, callbacks=[checkpoint_callback])
````

````python
'''
학습시 매 epoch에 해당하는 training set과 test set에 대한 손실값과 평가 매트릭스 수치를 확인할 수 있다.
이를 시각화하면 오버피팅 여부를 한 눈에 확인할 수 있다.
'''
fig, loss_ax = plt.subplots()

acc_ax = loss_ax.twinx()

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'r', label='val loss')
loss_ax.set_ylim([-0.2, 1.2])

acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')
acc_ax.plot(hist.history['val_accuracy'], 'g', label='val acc')


loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuracy')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
````

````python
'''
model을 로딩하여 test set으로 최종 평가
'''
model = load_model('best_model.h5')
loss_and_metrics = model.evaluate(x_test, y_test, batch_size=64)
print(loss_and_metrics)
````

### 순환 신경망 모델
문장을 단어들의 시퀀스로 간주하고 순환(LSTM) 레이어의 입력으로 구성한 모델이다. 임베딩 레이어 다음에 LSTM 레이어가 오는 경우에는 임베딩 레이어에 input_length 인자를 따로 설정할 필요는 없다. 입력 문장의 길이에 따라 input_length가 자동으로 정해지고 이것이 LSTM 레이어에는 timesteps으로 입력되기 때문이다. 블록으로 표현한다면 예제에서는 문장의 길이가 200 단어이므로 LSTM 블록 200개가 이어졌다고 생각하면 된다.

````python
'''
1. 네트워크 아키텍쳐 정의
'''
model = Sequential()
model.add(Embedding(20000, 128))
model.add(LSTM(128))
model.add(Dense(1, activation='sigmoid'))
````

````python
'''
2. 컴파일
3. 학습
'''
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

checkpoint_callback = ModelCheckpoint('best_model.h5', save_best_only=True, monitor='val_loss')

hist = model.fit(x_train, y_train, epochs=5, batch_size=64, validation_split=0.2, callbacks=[checkpoint_callback])
````

````python
%matplotlib inline

fig, loss_ax = plt.subplots()

acc_ax = loss_ax.twinx()

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'r', label='val loss')
loss_ax.set_ylim([-0.2, 1.2])

acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')
acc_ax.plot(hist.history['val_accuracy'], 'g', label='val acc')


loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuracy')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()
````

````python
model = load_model('best_model.h5')
loss_and_metrics = model.evaluate(x_test, y_test, batch_size=64)
print(loss_and_metrics)
````
